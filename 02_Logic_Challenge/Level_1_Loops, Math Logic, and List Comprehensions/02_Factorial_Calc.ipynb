**Write a Function to calculate factorial of a number**

```python
def fact():
  X = int(input("Enter Your Number : "))
  fact = 1
  for i in range(1,X+1):
    fact = fact*i
  print("The Factorial of %d = %d"%(X,fact))
fact()
```
