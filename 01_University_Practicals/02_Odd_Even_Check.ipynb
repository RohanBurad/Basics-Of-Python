**Implement a program to check whether a number is even or odd using conditional**

```python
#We will take numbers from the user to check the number is odd or even
number = float(input("Enter the number to be checked : "))
if number < 0:
  print("Invalid check ! Numbers can't be negative for odd / even check.")
else:
  if number % 2 == 0:
    print(f"The {number} is Even Number.")
  else:
    print(f"The {number} is Odd Number.")
```
